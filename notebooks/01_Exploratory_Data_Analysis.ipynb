# Exploratory Data Analysis & Flight Envelopes
This notebook explores the Hypersonic Physics Dataset, focusing on distributions and the safe flight envelope.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configure styling
sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
data = pd.read_csv('../hypersonic_dataset_138600pts.csv')
data.head()
```

## Dataset Overview
Let's check the basic statistics and data types.

```python
data.info()
```

```python
data.describe()
```

## Flight Envelope (Mach vs Altitude)

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='Mach', y='Altitude_km', hue='Thermal_Regime', style='Survived', alpha=0.7)
plt.title('Hypersonic Flight Envelope: Mach vs Altitude')
plt.xlabel('Mach Number')
plt.ylabel('Altitude (km)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## TPS Material Distribution

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Material', hue='Survived')
plt.title('TPS Material Survival Distribution')
plt.show()
```

## Correlation Heatmap

```python
numeric_data = data.select_dtypes(include=[np.number])
corr = numeric_data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

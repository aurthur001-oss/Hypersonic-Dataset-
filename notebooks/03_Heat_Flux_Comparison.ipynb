# Aerothermodynamics & Heat Flux Models Comparison
Comparing theoretical heat flux models: Fay-Riddell, Sutton-Graves, and Boundary Layer models.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
data = pd.read_csv('../hypersonic_dataset_138600pts.csv')
```

## Heat Flux Models Overview

```python
heat_flux_cols = ['q_FayRiddell_Wm2', 'q_SuttonGraves_Wm2', 'q_BoundaryLayer_Wm2']

plt.figure(figsize=(12, 6))
for col in heat_flux_cols:
    sns.kdeplot(data[col], log_scale=True, label=col, fill=True)
plt.title('Distribution of Heat Flux Predictions (Log Scale)')
plt.xlabel('Heat Flux (W/m²)')
plt.legend()
plt.show()
```

## Heat Flux vs Mach Number

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Mach', y='q_SuttonGraves_Wm2', label='Sutton-Graves')
sns.lineplot(data=data, x='Mach', y='q_FayRiddell_Wm2', label='Fay-Riddell')
plt.title('Stagnation Point Heat Flux vs Mach Number')
plt.ylabel('Heat Flux (W/m²)')
plt.yscale('log')
plt.show()
```

## Design Heat Flux vs Material Limits

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='q_w_design_Wm2', y='T_wall_outer_K', hue='Material', alpha=0.5)
plt.axhline(y=1900, color='r', linestyle='--', label='CarbonCarbon Limit (~1900K)')
plt.title('Design Heat Flux vs Wall Temperature by Material')
plt.xlabel('Design Heat Flux (W/m²)')
plt.ylabel('Outer Wall Temperature (K)')
plt.legend()
plt.show()
```

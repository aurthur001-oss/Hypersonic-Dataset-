# Active Cooling System & Material Optimization
Analyzing the performance of active H2 cooling systems and material limits.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../hypersonic_dataset_138600pts.csv')
```

## Cooling Efficiency vs Coolant Mass Flow

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='mdot_kgs', y='eta_cooling', hue='Material')
plt.title('Cooling Efficiency vs Coolant Mass Flow')
plt.xlabel('Mass Flow Rate (kg/s)')
plt.ylabel('Cooling Efficiency (eta)')
plt.show()
```

## Wall Temperature Constraints

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Material', y='T_wall_outer_K')
plt.title('Outer Wall Temperature Distribution by Material')
plt.ylabel('Outer Wall Temp (K)')
plt.show()
```

## Pressure Drop Penalty

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='eta_cooling', y='dP_channel_Pa', hue='mdot_kgs', palette='viridis')
plt.title('Cooling Efficiency vs Pressure Drop')
plt.xlabel('Cooling Efficiency')
plt.ylabel('Pressure Drop (Pa)')
plt.show()
```
